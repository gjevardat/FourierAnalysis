# Fourier Series

Any continuous periodic function of period $P$ (or continuously piece wise) can be decomposed in a infinite sum of cosine and sine functions

$$
\begin{equation}
    f(x) = \frac{A_0}{2} + \sum_{n=1}^\infty A_n \cos(\frac{2\pi nx}{P}) + B_n \sin(\frac{2\pi nx}{P})
\end{equation}
$$
Each of the component function from $1$ to $n$ is named  $n^{th}$ harmonic. The sum of all harmonics when $n$ goes to $\infty$ is $f$

$A_n$ and $B_n$ are called the Fourier coefficients and are defined as follow:

$$
\begin{equation}
    A_n = \frac{2}{P}\int_{0}^P f(x)cos(\frac{2\pi nx}{P})\\
    B_n = \frac{2}{P}\int_{0}^P f(x)sin(\frac{2\pi nx}{P})
\end{equation}
$$

All following examples illustrate the various concept by plotting:
* The function $f$ to approximate
* The Fourier series
* The Individual harmonics that are summed
* The Fourier coefficients
* The relative error between the Fourier series and $f$

## Example with a step function
Below are 2 examples of Fourier series to approximate the step function

```python
x,f = step_function(period=2,upperval=1,lowerval=0,num_periods=3)

Na = [100]
for i in Na:
    A, B, freqs, SUM, H,ERR = fourier_series(f,x,P=2,num_harmonics=i)
    plot_fourier(f,x,A,B,freqs,H,SUM,i,ERR)
```

## Example with a triangle function

```python
period = 1
x, dx, f = triangle_function(P = period, x0=1)

Na = [20]
for i in Na:
    A, B, freqs, SUM, H, ERR = fourier_series(f,x,period,i)
    plot_fourier(f,x,A,B,freqs,H,SUM,i,ERR)
```

```python
period = np.pi/2
x, dx, f = triangle_function(P = period, x0=0)

Na = [2,20]
for i in Na:
    A, B, freqs, SUM, H, ERR = fourier_series(f,x,period,i)
    plot_fourier(f,x,A,B,freqs,H,SUM,i,ERR)
```

## Example with a rectangle function

```python
period = 2
x, dx, f = square_function()
Na = [10]
for i in Na:
    A, B, freqs, SUM, H, ERR = fourier_series(f,x,period,i)
    plot_fourier(f,x,A,B,freqs,H,SUM,i,ERR)
```
