```python
import matplotlib.pyplot as plt
import numpy as np
import fourier

from fourier import *
```

```python
#step function period 4 upval 1 low
x,f = step_function(dx=0.01,period=3,upperval=1,lowerval=-1,num_periods=4)

Na = [3,50]
for i in Na:
    A, B, freqs, SUM, H = fourier_series(f,x,P=3,num_harmonics=i)
    plot_fourier(f,x,A,B,freqs,H,SUM,i)
```

```python
period = 1
x, dx, f = triangle_function(dx=0.01,P = period, x0=1)

Na = [0,1,20]
for i in Na:
    A, B, freqs, SUM, H = fourier_series(f,x,period,i)
    plot_fourier(f,x,A,B,freqs,H,SUM,i)
```

```python
period = np.pi/2
x, dx, f = triangle_function(dx=0.01,P = period, x0=0)

Na = [2,20]
for i in Na:
    A, B, freqs, SUM, H = fourier_series(f,x,period,i)
    plot_fourier(f,x,A,B,freqs,H,SUM,i)
```
